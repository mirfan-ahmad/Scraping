```python
from selenium.webdriver import Firefox
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException
from time import sleep
import os
```

```python
driver = Firefox()
driver.maximize_window()
```

```python
def scrape_data(url):
    driver.get(url)
    i = 1
    while True:
        try:
            anchor = driver.find_element(By.CSS_SELECTOR, f'div.infinite-item:nth-child({i}) > div:nth-child(2) > a:nth-child(2)')
            driver.get(anchor.get_attribute('href'))
            next_img = 1
            while True:
                try:
                    driver.find_element(By.CSS_SELECTOR, '#media-viewer-trigger > div.cover-img > img').click()
                    lnk = driver.find_element(By.ID, f'property-fs-img-{next_img}').get_attribute('src')
                    driver.get(lnk)
                    print(lnk)
                    driver.get_screenshot_as_file(f'/home/umair/Desktop/irfan/Scraping/Imgs/{lnk[:-20:-6]}.png')
                    next_img += 1
                    print('downladed')
                    driver.back()
                except NoSuchElementException:
                    print('Not found')
                    break
            driver.back()
            driver.back()
            i += 1
        except NoSuchElementException:
            print('Unable')
            i += 1
            return
```

```python
def main(url):
    j = 1
    while True:
        try:
            scrape_data(f'{url}?p={j}')
            j += 1
        except Exception as e:
            print('in except Block')
            print('error', str(e))
            return
```

```python
main('https://www.century21.com/real-estate/miami-fl/LCFLMIAMI/')
```
