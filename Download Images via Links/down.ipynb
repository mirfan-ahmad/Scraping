```python
import pandas as pd
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import os
```

```python
df = pd.read_excel('proptexx_sample_outgoing.xlsx')
df
```

```python
def scrap():
    links = df['mediaurl'].tolist()
    ser = Service('C:/Users/Irfan Ahmad/Downloads/chromedriver_win32/chromedriver.exe')
    driver = Chrome(service=ser)
    driver.maximize_window()
    directory_path = 'E:/Py/'
    directory_name = 'images'
    new_directory_path = os.path.join(directory_path, directory_name)
    os.makedirs(new_directory_path)
    os.chdir(new_directory_path)
    for i in range(len(links)):
        try:
            driver.get(links[i])
            url = driver.find_element(By.TAG_NAME, 'img').get_attribute('src')
            driver.get(url)
            with open(f'{i+1}', 'wb') as file:
                file.write(driver.find_element(By.TAG_NAME, 'img').screenshot_as_png)
        except:
            pass
        os.chdir(new_directory_path)

scrap()

```

```python
folder_path = '/home/irfan/Axcelerate AI/Machine Learning/Watermark Dataset/Text Watermark/'
file_list = os.listdir(folder_path)
i = 127
for file_name in file_list:
    original_file_path = os.path.join(folder_path, file_name)
    new_file_path = os.path.join(folder_path, f'{i}.png')
    os.rename(original_file_path, new_file_path)
    i += 1
```
